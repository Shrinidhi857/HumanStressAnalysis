```python
from pydub import AudioSegment
import os

def split_audio_in_chunks(audio_file_path, output_dir, chunk_length=2):
    audio = AudioSegment.from_wav(audio_file_path)
    chunk_length_ms = chunk_length * 60 * 1000  # 2 minutes = 120,000 milliseconds
    os.makedirs(output_dir, exist_ok=True)
    total_length = len(audio)  # Total audio duration in milliseconds
    num_chunks = (total_length + chunk_length_ms - 1) // chunk_length_ms  # Calculate number of chunks

   # print(f"Total audio duration: {total_length / 1000:.2f} seconds")
    #print(f"Number of chunks: {num_chunks}")

    for i in range(num_chunks):
        start_time = i * chunk_length_ms
        end_time = min((i + 1) * chunk_length_ms, total_length)  # Ensure no overflow
        chunk = audio[start_time:end_time]
        chunk_filename = os.path.join(output_dir, f"{audio_file_path[-13:-10]}_{i+1}.wav")
        chunk.export(chunk_filename, format="wav")
        print(f"Saved chunk {i+1} to {chunk_filename}")


```

```python
for file in tqdm(os.listdir(audio_dir)):
    if file.endswith(".wav"):
        file_path = os.path.join(audio_dir, file)
        split_audio_in_chunks(file_path, output_directory, chunk_length=2)
```
