**Merging the two dataframes **

```python
data1="/content/drive/MyDrive/Colab Notebooks/train_split_Depression_AVEC2017.csv"
data2="/content/drive/MyDrive/Colab Notebooks/dev_split_Depression_AVEC2017.csv"

```

```python
df1 = pd.read_csv(data1)
df2 = pd.read_csv(data2)


combined_df = pd.concat([df1, df2])


unique_rows_df = combined_df.drop_duplicates()
unique_rows_df = unique_rows_df.sort_values(by=unique_rows_df.columns[0]).reset_index(drop=True)
print("Unique Rows Sorted DataFrame:")
print(unique_rows_df)
```

```python
data_df=pd.DataFrame(columns=['id','file_path','depression_level'])
```

```python
for file in os.listdir(input_dir):
  file_path=os.path.join(input_dir,file)
  id=file[:3]
  try:
    depression_level=unique_rows_df[unique_rows_df['Participant_ID']==int(id)]['PHQ8_Score'].values[0]
  except:
    depression_level=0

  new_row = pd.DataFrame([[id, file_path, depression_level]], columns=["id", "file_path", "depression_level"])
  data_df = pd.concat([data_df, new_row], ignore_index=True)
```
