```python
# Function combining all of above
def remove_silent(participantID):

    ##############
    # Read + Normalize
    ##############



    fs, signal = wavfile.read(f"/content/drive/MyDrive/Colab Notebooks/audio-step1-diacWoz/{participantID}_AUDIO.wav")
    signal = signal / (2**15) # Normalize


    ##############
    # Segmentation
    ##############


    signal_len = len(signal)
    segment_size_t = 1 # segment size in seconds
    segment_size = segment_size_t * fs  # segment size in samples
    # Break signal into list of segments in a single-line Python code
    segments = np.array([signal[x:x + segment_size] for x in
                        np.arange(0, signal_len, segment_size)])
    # Create new folder to save segments
    # Specify the directory path you want to create
    directory_path = f'/content/drive/MyDrive/Colab Notebooks/audio-step3-silence-temprory/{participantID}'

    # Create the directory
    os.makedirs(directory_path, exist_ok=True)  # exist_ok=True prevents error if directory already exists

    print(f"Directory created at {directory_path}")


    for iS, s in enumerate(segments):
        wavfile.write(directory_path+"/{0}_Audio_segment_{1:d}_{2:d}.wav".format(participantID,segment_size_t *iS,segment_size_t * (iS + 1)), fs, (s))

    ##############
    # Removing silent segments
    ##############



    # Remove pauses using an energy threshold = 50% of the median energy:
    energies = [(s**2).sum() / len(s) for s in segments]
    # (attention: integer overflow would occure without normalization here!)
    thres = 0.5 * np.median(energies)
    index_of_segments_to_keep = (np.where(energies > thres)[0])
    # get segments that have energies higher than a the threshold:
    segments2 = segments[index_of_segments_to_keep]
    # concatenate segments to signal:
    new_signal = np.concatenate(segments2)
    # and write to file:


    wavfile.write(f"/content/drive/MyDrive/Colab Notebooks/audio-step3-silence-removed/{participantID}_AUDIO_p.wav", fs, new_signal)
    print("Participant ID: " + str(particpantID) + " DONE!")


```

```python
for file_path in tqdm(glob.glob(os.path.join(input_dir, "*.wav"))):
    if int(os.path.basename(file_path)[0:3]) in problem:
        continue  # Skip files with prefixes in the 'problem' list

    try:
        print(os.path.basename(file_path)[0:3])
        remove_silent(os.path.basename(file_path)[0:3])
    except Exception as e:
        print(f"Error processing file {file_path}: {e}")
```
