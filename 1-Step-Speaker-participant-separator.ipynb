```python
from pydub import AudioSegment
import os

def get_audio_chunk(audio, start_time, end_time):
    start_ms = int(start_time * 1000)
    end_ms = int(end_time * 1000)
    return audio[start_ms:end_ms]
```

```python
def combine_chunks(chunk_files, output_file):
    combined = AudioSegment.empty()
    for chunk_file in chunk_files:
        chunk = AudioSegment.from_wav(chunk_file)
        combined += chunk

    combined.export(output_file, format="wav")
    print(f"Combined audio saved: {output_file}")

```

```python
from tqdm import tqdm
```

```python
for file in tqdm(file_list):
  print(file[:3])
  audio_path = os.path.join(Audio_dir, file)
  audio = AudioSegment.from_wav(audio_path)

  # Initialize an empty combined audio segment
  combined_audio = AudioSegment.empty()

  # Load transcript CSV
  transcript_file = os.path.join(trans_dir, file[:3] + "_TRANSCRIPT.csv")
  trans_df = pd.read_csv(transcript_file, sep="\t")
  print("Column names in CSV:", trans_df.columns)

  # Loop through the transcript rows and extract Participant audio chunks
  for i in range(len(trans_df)):
      if trans_df["speaker"][i] == 'Participant':
          start_time = trans_df["start_time"][i]
          end_time = trans_df["stop_time"][i]  # Correct column name
          chunk = get_audio_chunk(audio, start_time, end_time)
          combined_audio += chunk

  # Export the combined audio for this file
  output_file_path = os.path.join(output_dir, file)
  combined_audio.export(output_file_path, format="wav")
  print(f"Saved combined audio to: {output_file_path}")
```
